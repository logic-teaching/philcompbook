*Note 2*: for a function $f(x,y,z)$ with three inputs, the [continuity conditions](https://logic-teaching.github.io/philcompbook/Chap10.html#the-continuous-functions) are:

1. *Monotonicity*: $d_1\sqsubseteq d_1^{\prime}, d_2\sqsubseteq d_2^{\prime}, d_3\sqsubseteq d_3^{\prime}$ implies $f(d_1, d_2, d_3)\sqsubseteq f(d_1^{\prime}, d_2^{\prime}, d_3^{\prime})$
2. *Continuity*: $d_{1}\sqsubseteq d_{2}\sqsubseteq d_{3}\sqsubseteq \cdots$ and $d_{1}^{\prime}\sqsubseteq d_{2}^{\prime}\sqsubseteq d_{3}^{\prime}\sqsubseteq \cdots$ and $d_{1}^{\prime\prime}\sqsubseteq d_{2}^{\prime\prime}\sqsubseteq d_{3}^{\prime\prime}\sqsubseteq \cdots$ implies $\bigsqcup_{n>0} f(d_{n}, d_n^{\prime}, d_n^{\prime\prime})=f(\bigsqcup_{n>0} d_{n}, \bigsqcup_{n>0} d_{n}^{\prime}, \bigsqcup_{n>0} d_{n}^{\prime\prime})$
